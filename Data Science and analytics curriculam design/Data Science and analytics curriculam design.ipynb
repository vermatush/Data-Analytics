##2.1) Cleaning courses data

```python
# Arranging the dataframe df_scrapped into a one-hot encoding data where if the skill is present in the
# description we will put one in under the skill column else we will put zero

skill_dataframe_dict = {}

a_string = "abc !? 123-"
for index, row in raw_courses.iterrows():
    
    title = str(row['Courses'])
    description = str(row['Syllabus'])
    description = description.lower()
    #print(description)
    #break
    result = {} 
    for skill_category, skills in skills_keywords_dict.items():
        
        # variable used to store results of the intermediate check (loop below)
        category_found = 0   
               
        # loop over all skills in the sublist of 'skills_keywords_dict'
        for skill in skills:
            skill = skill.lower()
                    
            # if skill from the sublist is found, set 'category_found' to 1
            if skill.lower() in description.lower():
                category_found = 1
            if description.find(skill) != -1:
                category_found = 1                     
            
            # skill set to 1 if found, 0 if not, in 'results_dict'
            result[skill_category] = category_found 
    skill_dataframe_dict[title] = result
    
skills_encoded_df_courses = pd.DataFrame(skill_dataframe_dict).T.reset_index()
skills_encoded_df_courses.head(5)
```

```python
# Arranging the dataframe df_scrapped into a one-hot encoding data where if the skill is present in the
# description we will put one in under the skill column else we will put zero

skill_dataframe_dict = {}

a_string = "abc !? 123-"
for index, row in df.iterrows():
    
    title = str(row['Q5'])
    description = str(row['skills'])
    description = description.lower()
    #print(description)
    #break
    result = {} 
    for skill_category, skills in skills_keywords_dict.items():
        
        # variable used to store results of the intermediate check (loop below)
        category_found = 0   
            
        # loop over all skills in the sublist of 'skills_keywords_dict'
        for skill in skills:
            skill = skill.lower()
            # if skill from the sublist is found, set 'category_found' to 1
            if skill.lower() in description.lower():
                category_found = 1
            if description.find(skill) != -1:
                category_found = 1                     
            # skill set to 1 if found, 0 if not, in 'results_dict'
            result[skill_category] = category_found 
    skill_dataframe_dict[title] = result
skills_encoded_df = pd.DataFrame(skill_dataframe_dict).T.reset_index()
skills_encoded_df
```

```python
init_data = data

def get_average_from_str(x):
    empty_check = [' years','+','<','>','$',',']
    
    for ec in empty_check:
        if(ec in str(x)):
            x = str(x).replace(ec, '')
            
    if('I have never written code' in str(x)):
        x = x.replace('I have never written code', '0')
        
    if('-' in str(x)):
        x = x.split('-')
        x = ( int(x[0]) + int(x[1]) ) / 2
        
    return float(x)

# get the single average value from the salary range column
def average_salary(df_series,column):
    salary_ave = []
    
    for gs in df_series:
        a = round( init_data.loc[init_data[column] == gs]['average_salary'].mean(), 2 )
        salary_ave.append(a)
    return salary_ave

def change_order(_series,_values):
    
    new_df = pd.DataFrame()
    new_df['series'] = _series
    new_df['values'] = _values
    new_df = new_df.sort_values(by = 'values', ascending = False)
    return {'series': new_df['series'].to_list(), 'values': new_df['values'].to_list()}

# Gender
gender_series = pd.Series(init_data.Q2).drop_duplicates().to_list()
gender_count = []
for s in gender_series:
    gender_count.append(init_data['Q2'].str.count(s).sum() )

# Country
country_series = pd.Series(init_data.Q3).drop_duplicates().to_list()
country_count = []
for s in country_series:
    country_count.append( init_data['Q3'].str.count(s).sum() )
    
new_country_order = change_order(country_series,country_count)

# Age
age_series = pd.Series(init_data.Q1).drop_duplicates().to_list()
age_count = []
for s in age_series:
    age_count.append( init_data['Q1'].str.count(s).sum() )

# Education
education_series = pd.Series(init_data.Q4).drop_duplicates().to_list()
education_count = []
for s in education_series:
    education_count.append( init_data['Q4'].str.count(s).sum() )
  
# Recent Role Demographic
role_series = pd.Series(init_data.Q5).drop_duplicates().to_list()
role_count = []
for s in role_series:
    role_count.append( init_data['Q5'].str.count(s).sum() )

new_role_order = change_order(role_series,role_count)

# Programming Experience Demographic
prog_series = pd.Series(init_data.Q6).drop_duplicates().to_list()
prog_count = []
for s in prog_series:
    prog_count.append( init_data['Q6'].str.count(s).sum() )

# Creating Column average_salary
init_data['average_salary'] = init_data['Q25'].apply(get_average_from_str)
```

```python
# Get average salary per role
role_salary_ave = average_salary(role_series,'Q5')

new_role_order = change_order(role_series,role_salary_ave)
           
fig = plt.figure(figsize=(10,5), dpi=100)
fig.suptitle('Average Salary per Role',fontsize='10')

ax = fig.add_axes([0,0,1,1])
#ax.barh(role_series,role_salary_ave,color=['#003f5c'])
ax.barh(new_role_order['series'],new_role_order['values'],color='#003E7A')

# Show top values 
ax.invert_yaxis()
    
# Add x,y gridlines
ax.grid(b=True, color='grey', linestyle='-.', linewidth=0.5, alpha=0.2)

# Add annotation to bars
for i in ax.patches:
    ax.text(i.get_width()+500, i.get_y()+0.5, str(round((i.get_width()), 2)), fontsize=10, fontweight='bold', color='grey')

ax.set_xlabel('Salary')
fig.show()
```
